```python
import pandas as pd # This library is to enable me to transform the JSON data into a dataframe
import requests # Importing the library needed to read the JSON Data link to be used
```

```python
# Assigning my JSON data raw github link to an identifier(url) 
url = 'https://raw.githubusercontent.com/DerickMenje/LuxDevHq_Weekly_Assignments/refs/heads/main/Working%20with%20JSON%20Data%20From%20Github/MOCK_DATA.json'
```

```python
# Sending the GET request
response = requests.get(url)
```

```python
# Parsing the JSON content into a dataframe
data = response.json()
```

```python
# Converting the JSON into a dataframe
df = pd.Dataframe(data)
```

```python
# Saving the data frame as a CSV
df.to_csv('MyMock_Up.csv', index = False)

# index = False ensures that the csv does not add another column of row number
```
