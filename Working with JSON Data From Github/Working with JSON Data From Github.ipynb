## Working with JSON Data from GitHub

```python
import pandas as pd # This library is to enable me to transform the JSON data into a dataframe
import requests # Importing the library needed to read the JSON Data link to be used
```

```python
# Assigning my JSON data raw github link to an identifier(url) 
url = 'https://raw.githubusercontent.com/DerickMenje/LuxDevHq_Weekly_Assignments/refs/heads/main/Working%20with%20JSON%20Data%20From%20Github/MOCK_DATA.json'
```

```python
# Sending the GET request
response = requests.get(url)
```

```python
# Parsing the JSON content into a dataframe
data = response.json()
```

```python
# Converting the JSON into a dataframe
df = pd.DataFrame(data)
df.head()
```

```python
# Saving the data frame as a CSV
df.to_csv('MyMock_Up.csv', index = False)

# index = False ensures that the csv does not add another column of row number
```

## Working with Dummy JSON API Endpoints

```python
url_1 = 'https://dummyjson.com/products'
url_2 = 'https://dummyjson.com/carts'
```

```python
response_1 = requests.get(url_1)
response_2 = requests.get(url_2)

data_1 = response_1.json()
data_2 = response_2.json()


# Flattening the Products and carts lists into a table; This breaks down the items in the product into columns.
df_products = pd.json_normalize(data_1['products'])
df_carts = pd.json_normalize(data_2['carts'])

# Converting to CSV
df_products.to_csv('Products.csv', index = False)
df_carts.to_csv('carts.csv', index = False)
```
